```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE, MDS, Isomap
```

```python
def get_sphere(radius : int = 5, sample_n : int = 10000) -> tuple:
    theta = radius*np.random.randn(sample_n)*np.pi
    pi = radius*np.random.randn(sample_n)*2*np.pi
    x = np.random.randint(-radius//2, radius//2, sample_n) + radius*np.sin(theta)*np.cos(pi)
    y = np.random.randint(-radius//2, radius//2, sample_n) + radius*np.sin(theta)*np.sin(pi)
    z = np.random.randint(-radius//2, radius//2, sample_n) + radius*np.cos(theta)

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    ax.scatter(x, y, z, alpha = 0.5)
    plt.show()

    return x, y, z

x, y, z = get_sphere()
```

```python
def exc_PCA(x : np.ndarray, y : np.ndarray, z : np.ndarray) -> np.ndarray:
    X = np.stack((x, y, z), axis=-1)
    pca = PCA(n_components=2)
    X2D = pca.fit_transform(X)

    after_x = X2D[:,0]
    after_y = X2D[:,1]
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(after_x, after_y, alpha = 0.5)
    plt.show()
    return X2D

exc_PCA(x, y, z)
    
```

```python
def exc_tSNE(x : np.ndarray, y : np.ndarray, z : np.ndarray) -> np.ndarray:
    X = np.stack((x, y, z), axis=-1)
    tSNE = TSNE(n_components=2)
    X2D = tSNE.fit_transform(X)

    after_x = X2D[:,0]
    after_y = X2D[:,1]
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(after_x, after_y, alpha = 0.5)
    plt.show()
    return X2D

exc_tSNE(x, y, z)
    
```

```python
def exc_isomap(x : np.ndarray, y : np.ndarray, z : np.ndarray) -> np.ndarray:
    X = np.stack((x, y, z), axis=-1)
    isomap = Isomap(n_components=2)
    X2D = isomap.fit_transform(X)

    after_x = X2D[:,0]
    after_y = X2D[:,1]
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(after_x, after_y, alpha = 0.5)
    plt.show()
    return X2D

exc_isomap(x, y, z)
```
